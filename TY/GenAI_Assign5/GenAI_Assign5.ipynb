```python
import torch
from diffusers import StableDiffusionPipeline
```

```python
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

```python
prompt = "A dreamlike ocean scene with transparent waves, floating jellyfish emitting soft light, underwater ruins, and a moonlit sky merging with the sea"

```

```python
from IPython.display import display

image = pipe(prompt).images[0]
display(image)
```
